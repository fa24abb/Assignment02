### Problem 01

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 300 random data points using uniform distribution
low = 3
high = 5
data = np.random.uniform(low=low, high=high, size=300)
```

```python
# histogram of the data
plt.figure(figsize=(10, 6))  # Set figure size for better readability
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)

# Mean and Median
mean = np.mean(data)
median = np.median(data)

# annotations for Mean
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.text(mean + 0.05, 25, f'Mean: {mean:.2f}', color='red', fontsize=12)

# annotations for Median
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median:.2f}')
plt.text(median - 0.2, 25, f'Median: {median:.2f}', color='green', fontsize=12)

# appearance for readability
plt.title("Histogram of Uniform Distribution Data", fontsize=16)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.legend(loc='upper right', fontsize=12)
plt.tick_params(axis='both', labelsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

### Problem 02


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pyarrow.parquet as pq

# Load Parquet file from  URL
url = "https://star.herts.ac.uk/~kuhn/DHV/exercises_problem2.parquet"
df = pd.read_parquet(url)

# checking what is in dataset
df.head()
df.info()
df.describe()
```

```python
# missing values checking
df.isnull().sum()
```

```python
# figure size setting
plt.figure(figsize=(15, 5))

# Plot histogram for var1
plt.subplot(1, 3, 1)
sns.histplot(df['var1'], kde=True, color='skyblue')
plt.title('Distribution of var1', fontsize=16)
plt.xlabel('var1', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Plot histogram for var2
plt.subplot(1, 3, 2)
sns.histplot(df['var2'], kde=True, color='lightgreen')
plt.title('Distribution of var2', fontsize=16)
plt.xlabel('var2', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Plot histogram for var3
plt.subplot(1, 3, 3)
sns.histplot(df['var3'], kde=True, color='salmon')
plt.title('Distribution of var3', fontsize=16)
plt.xlabel('var3', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# figure size setting
plt.figure(figsize=(15, 5))

# Boxplot for var1 with hue set to 'class' column 
plt.subplot(1, 3, 1)
sns.boxplot(x='class', y='var1', data=df, hue='class', palette='Set2', showfliers=False)
plt.title('Boxplot of var1 by class', fontsize=16)
plt.xlabel('Class', fontsize=14)
plt.ylabel('var1', fontsize=14)

# Boxplot for var2 with hue set to 'class' column
plt.subplot(1, 3, 2)
sns.boxplot(x='class', y='var2', data=df, hue='class', palette='Set2', showfliers=False)
plt.title('Boxplot of var2 by class', fontsize=16)
plt.xlabel('Class', fontsize=14)
plt.ylabel('var2', fontsize=14)

# Boxplot for var3 with hue set to 'class' column
plt.subplot(1, 3, 3)
sns.boxplot(x='class', y='var3', data=df, hue='class', palette='Set2', showfliers=False)
plt.title('Boxplot of var3 by class', fontsize=16)
plt.xlabel('Class', fontsize=14)
plt.ylabel('var3', fontsize=14)

# Adjust the layout for better spacing
plt.tight_layout()
plt.show()
```

```python
# Pairplot for numeric variables, colored by class column
sns.pairplot(df[['var1', 'var2', 'var3', 'class']], hue='class', palette='Set1')
plt.suptitle('Pairplot of var1, var2, var3 by class', fontsize=16)
plt.show()
```

```python
# Compute the correlation matrix for the numeric variables
corr = df[['var1', 'var2', 'var3']].corr()

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
